# Stanford admissions trends, 2008-2026

Charts the 18-year evolution of:
- application volume
- admit rate (admitted / applied)
- yield (enrolled / admitted)
- median test scores
- sticker tuition vs. average financial aid package

```python
import pandas as pd
import matplotlib.pyplot as plt

DATA = '../data/csv'

adm = pd.read_csv(f'{DATA}/admissions_summary.csv')
adm = adm.pivot(index='year', columns='metric', values='value').sort_index()
adm['admit_rate']  = adm['admitted'] / adm['applied']
adm['yield_rate']  = adm['enrolled'] / adm['admitted']
adm
```

```python
fig, ax = plt.subplots(2, 2, figsize=(11, 7))
adm['applied'].plot(ax=ax[0,0], marker='o', title='Applications received')
(adm['admit_rate']*100).plot(ax=ax[0,1], marker='o', color='#990000', title='Admit rate (%)')
(adm['yield_rate']*100).plot(ax=ax[1,0], marker='o', color='#005500', title='Yield (%)')
adm['enrolled'].plot(ax=ax[1,1], marker='o', title='Enrolled')
for a in ax.flatten():
    a.tick_params(axis='x', rotation=45)
plt.tight_layout()
```

```python
ts = pd.read_csv(f'{DATA}/test_scores.csv')
sat_total = ts[ts.test == 'SAT_total'].set_index('year').sort_index()
sat_total[['p25', 'p50', 'p75']].plot(
    figsize=(9, 4), marker='o', title='SAT total — 25th / 50th / 75th percentile'
)
plt.xticks(rotation=45);
```

```python
tu = pd.read_csv(f'{DATA}/tuition_and_fees.csv')
tu = tu.pivot(index='year', columns='expense', values='amount').sort_index()
fa = pd.read_csv(f'{DATA}/financial_aid_summary.csv')
fa = fa.pivot(index='year', columns='metric', values='value').sort_index()
compare = pd.concat([tu['tuition_first_year'].rename('tuition'),
                     fa['avg_financial_aid_package'].rename('avg_aid_package')], axis=1)
compare.plot(figsize=(9, 4), marker='o', title='Sticker tuition vs. average aid package')
plt.ylabel('USD'); plt.xticks(rotation=45);
```
